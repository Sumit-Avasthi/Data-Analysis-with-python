## Pandas Library

```python
import pandas as pd
```

```python
# Series
#Using 1-D list (default indexes 0,1,2,3...) and (values -> collection el)
data = [5,6,7,8,9,10]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
#Series using dictionary
dic = {'a':1,'b':2,'c':3}  
series_dict = pd.Series(dic)  #(key->indexes) and (values->series value)
print(series_dict)
print(type(series_dict))
```

```python
#Creating Series with Our indexes

lst = [10,11,12,13,14]
indx = ['a','b','c','d','e']
print(pd.Series(lst,index=indx))
```

```python
## Dataframe in pandas(2-D dimensional)-Data like in excel or csv sheet

## Creating DataFrame from Dictionary

dic = {
    'Name':['Sumit','Shubham','Amit','Suryansh'],
    'Age':[20,21,24,22],
    'City':['Kanpur','Kanpur','Kanpur','Roora'],
    'Grade':['A','B+','A+','B'],
}

df = pd.DataFrame(dic)
print(df)
print(type(df))
```

```python
import numpy as np

print(np.array(df))
```

```python
# Using List of dictionary

lst = [
    {'Name':'Sumit','Age':20,'City':'Kanpur','Grade':'A'},
    {'Name':'Shubham','Age':22,'City':'Kanpur','Grade':'B'},
    {'Name':'Suryansh','Age':23,'City':'Kanpur','Grade':'B+'},
    {'Name':'Amit','Age':24,'City':'Kanpur','Grade':'A+'},
]

df = pd.DataFrame(lst)
print(df)
```
