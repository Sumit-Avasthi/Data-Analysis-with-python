### Basics of MatplotLib

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

# To create a line plot

plt.plot(x,y) # To create plot
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Basic Line Plot')
plt.show() # To show our plot 
```

```python
# Custome Line plot
plt.plot(x,y,color="red",linestyle="--",marker="o",linewidth=3,markersize=9)
plt.grid(True)
plt.show()
```

```python
# To Plot Multiple subplot in a one figure
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]
plt.figure(figsize=(9,5))


# 1 subplot
plt.subplot(1,3,1) # 1-row 2-column(2 plots) 1st plot
plt.plot(x,y1,color="green")
plt.title('Plot 1st')

plt.subplot(1,3,2) # 1-row 2-column 1st plot
plt.plot(x,y2,color="red")
plt.title('Plot 2st')

plt.subplot(1,3,3) # 1-row 2-column 1st plot
plt.plot(x,y2,color="blue")
plt.title('Plot 3st')
plt.show()
```

```python
## Bar Plot

categories=['A','B','C','D','E']
values=[4,10,7,2,6]

plt.bar(categories,values,color="purple",edgecolor="red")
plt.show()
```

```python
## Histogram - shows distribution of dataset. it divide the data in bins and count the number of data points in each bins
data = [1,1,2,3,4,5,5,6,7,7,8,8,4,2,3,5,6,4,7,7,8,8,8,6,43,2,1,1]

plt.hist(data,bins=8,color="orange",edgecolor="black")
plt.show()
```

```python
## Scatter plot
x = [1,2,3,4,5]
y=[1,2,3,4,5]
plt.scatter(x,y,color="red")
plt.show()
```

```python
# pie plot

labels = ['A','B','C','D']
sizes=[30,20,40,10]
colors=['red','yellow','green','blue']
explod=[0.2,0,0,0] # move out the first slice
plt.pie(sizes,labels=labels,colors=colors,shadow=True,autopct='%1.1f%%',explode=explod)
plt.show()
```
